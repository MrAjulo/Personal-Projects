# Active Corona virus cases in NIgeria as of 20th of May 2020

This project help to visualize the number of active cases of the COvid 19 virus in Nigeria across the 36 states.


```
#Import Library
import folium
```

```
#Create Map of Nigeria
Nigeria_map = folium.Map(location=[9.0820, 8.6753], zoom_start = 7)

Nigeria_map
```

```
#Import Libraries
import pandas as pd
import numpy as np
```

```
#Import Libraries for Web Scrapping
import requests 
from bs4 import BeautifulSoup
from tabulate import tabulate
```

```
#Scrap data from NCDC website
res = requests.get('https://covid19.ncdc.gov.ng/')
```

```
soup = BeautifulSoup(res.content, 'lxml')
```

```
table = soup.find_all('table')[0]
```

```
DF = pd.read_html(str(table))[0]
```

```
DF
```

```
#The cells below aim to convert each column in the NCDC dataset into a list 
States_Affected = DF["States Affected"].tolist()
```

```
Cases = DF['No. of Cases (Lab Confirmed)'].tolist()
```

```
Discharged = DF['No. Discharged'].tolist()
```

```
Deaths = DF['No. of Deaths'].tolist()
```

```
#This is to convert the lists above into a tuple to make it easier to convert to a dataframe
zippedList =  list(zip(States_Affected, Cases, Discharged, Deaths))
```

```
#The Dataframe Cov_Nig which contains list of states and their Corona virus cases
Cov_Nig = pd.DataFrame(zippedList, columns = ['States' , 'Cases', 'Discharged', 'Deaths'])
```

```
#Now we merge the two dataframes together using the States column
df_merge_col = pd.merge(new, Cov_Nig, on='States')
df_merge_col
```

```
#Inputing the Abuja row
new_row = {'lat':7.3986, 'lng':9.0765, 'States':'Abuja', 'capital':'admin' ,'Cases':427, 'Discharged': 108, 'Deaths':9}
```

```
df_merge_col = df_merge_col.append(new_row, ignore_index=True)
```

```
df_merge_col
```

```
#Inputing the Kogi row
new_row = {'lat':6.6906, 'lng':7.7337, 'States':'Kogi', 'capital':'admin' ,'Cases':0, 'Discharged': 0, 'Deaths':0}
```

```
df_merge_col = df_merge_col.append(new_row, ignore_index=True)
df_merge_col.shape
```

```
# instantiate a feature group for the cases in the dataframe
Active = folium.map.FeatureGroup()
```

```
# loop through the df_merge_col and add each to the cases feature group
for lat, lng, in zip(df_merge_col.lat, df_merge_col.lng):
    Active.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=4, 
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
```

```
# add pop-up text to each marker on the map
latitudes = list(df_merge_col.lat)
longitudes = list(df_merge_col.lng)
labels = list(df_merge_col.Cases)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng],
                  popup= folium.Popup('<strong>' + str(label) + '</strong>')).add_to(Nigeria_map) 
       
    
# add Cases to map
Nigeria_map.add_child(Active)
```
